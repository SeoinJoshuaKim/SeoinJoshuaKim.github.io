  - Ïù¥ÌõÑ ÏïÑÎûò ÎßÅÌÅ¨Î•º Ï∞∏Ï°∞ÌïòÏó¨ Ï≤´Î≤àÏß∏ rowÎ•º column nameÏúºÎ°ú ÏÑ§Ï†ïÌïòÏÑ∏Ïöî.

<https://stackoverflow.com/questions/31328861/python-pandas-replacing-header-with-top-row>

- ÌååÏùºÏùÄ 2Í∞úÏùò ÏãúÌä∏Î°ú Íµ¨ÏÑ±ÎêòÏñ¥ ÏûàÏúºÎ©∞ Í∞ÅÍ∞Å Ïù¥Î¶ÑÏùÄ `008770`, `035250` ÏûÖÎãàÎã§.



````python
import pandas as pd
import numpy as np


data_url1 = 'https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/stocks/Travel.xlsx'

df1 = pd.read_excel(data_url1, '008770 ')

df1 = df1.transpose()

new_header = df1.iloc[0] 
df1 = df1[1:] #take the data less the header row
df1.columns = new_header #set the header row as the df header

df2 = pd.read_excel(data_url1,'035250')

df2 = df2.transpose()

new_header = df2.iloc[0] 
df2 = df2[1:] #take the data less the header row
df2.columns = new_header #set the header row as the df header
````
